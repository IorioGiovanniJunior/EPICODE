```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# ANALISI MELLUSO

```python
melluso = pd.read_csv("puliti/melluso.csv")

# MEDIA,MASSIMO E MINIMO PER OGNI CATEGORIA:
stats = melluso.groupby('categoria')['prezzo'].agg(['mean', 'max', 'min']).reset_index()

# DIMENSIONI DELLA FIGURA
plt.figure(figsize=(12, 6))

# PLOT DELLA MEDIA
plt.bar(stats['categoria'], stats['mean'], color='skyblue', label='Media')

# AGGIUNGIAMO LE LINEE PER MIN E MASSIMO
plt.plot(stats['categoria'], stats['max'], color='red', marker='o', label='Massimo')
plt.plot(stats['categoria'], stats['min'], color='green', marker='o', label='Minimo')

# PERSONALIZZATO
plt.xlabel('Categoria')
plt.ylabel('Prezzo')
plt.title('Media, Massimo e Minimo del Prezzo per Categoria di Melluso')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
# Salva il grafico come file PNG
plt.savefig('prezzo melluso nuovo.png')

# Opzionale: mostra il grafico a schermo
plt.show()

# MOSTRA IL GRAFICO:
plt.show()
```

```python
# CALCOLA MEDIA,MIN E MASSIMO PER OGNI MATERIALE
stats = melluso.groupby("mat1")['prezzo'].agg(['mean', 'max', 'min']).reset_index()

# DIMENSIONI IMMAGINE
plt.figure(figsize=(12, 6))

# PLOT MEDIA
plt.bar(stats["mat1"], stats['mean'], color='skyblue', label='Media')

# LINEE MAX/MIN
plt.plot(stats["mat1"], stats['max'], color='red', marker='o', label='Massimo')
plt.plot(stats["mat1"], stats['min'], color='green', marker='o', label='Minimo')

# PERSONALIZZATO
plt.xlabel('Materiale')
plt.ylabel('Prezzo')
plt.title('Media, Massimo e Minimo del Prezzo per Materiale nei prodotti di Melluso')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()


# MOSTRA
plt.show()
```

```python
# Calcola la media, il massimo e il minimo del prezzo per ogni colore
stats = melluso.groupby("colore1")['prezzo'].agg(['mean', 'max', 'min']).reset_index()

# Imposta la dimensione della figura
plt.figure(figsize=(12, 6))

# Plot della media
plt.bar(stats["colore1"], stats['mean'], color='skyblue', label='Media')

# Aggiungi le linee per il massimo e il minimo
plt.plot(stats["colore1"], stats['max'], color='red', marker='o', label='Massimo')
plt.plot(stats["colore1"], stats['min'], color='green', marker='o', label='Minimo')

# Personalizza il grafico
plt.xlabel('Colore')
plt.ylabel('Prezzo')
plt.title('Media, Massimo e Minimo del Prezzo per Colore nei prodotti di Melluso')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

# Mostra il grafico
plt.show()
```

```python
# Calcola le statistiche
stats = melluso.groupby("colore1")['prezzo'].agg(['mean', 'max', 'min']).reset_index()

# Ordina per media del prezzo
stats = stats.sort_values('mean')

# Imposta la dimensione della figura
plt.figure(figsize=(12, 6))

# Plot della media come barre
plt.bar(stats["colore1"], stats['mean'], color='skyblue', label='Media')

# Aggiungi le linee per il massimo e il minimo
plt.plot(stats["colore1"], stats['max'], color='red', marker='o', label='Massimo')
plt.plot(stats["colore1"], stats['min'], color='green', marker='o', label='Minimo')

# Personalizza il grafico
plt.xlabel('Colore')
plt.ylabel('Prezzo')
plt.title('Media, Massimo e Minimo del Prezzo per Colore (Ordinato per Media)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

plt.show()
```

```python
# Creiamo una tabella di contingenza (pivot table) con le categorie e i generi
conteggio = pd.crosstab(melluso['categoria'], melluso['genere'])

# Plot a barre affiancate
conteggio.plot(kind='bar', figsize=(12, 6), colormap='tab10')

plt.title('Numero di prodotti per Categoria e Genere di Melluso')
plt.xlabel('Categoria')
plt.ylabel('Numero di prodotti')
plt.xticks(rotation=45)
plt.legend(title='Genere')
plt.tight_layout()
# Salva il grafico come file PNG
plt.savefig('prodotti melluso per categoria2.png')

# Opzionale: mostra il grafico a schermo
plt.show()

plt.show()
```

```python
# Creiamo una tabella di contingenza (pivot table) con le categorie e i generi
conteggio = pd.crosstab(melluso['colore1'], melluso['genere'])

# Sommando le occorrenze per ogni colore
totali_per_colore = conteggio.sum(axis=1)

# Ordinare le categorie di colore in base alla somma totale, decrescente
top_25_colori = totali_per_colore.sort_values(ascending=False).head(25).index

# Filtrare il conteggio per mantenere solo le prime 25 categorie di colore
conteggio_top25 = conteggio.loc[top_25_colori]

# Plot a barre affiancate con le prime 25 categorie di colore ordinate
conteggio_top25.plot(kind='bar', figsize=(12, 6), colormap='tab10')

plt.title('Numero di prodotti per le prime 25 categorie di Colore e Genere di Melluso')
plt.xlabel('Colore')
plt.ylabel('Numero di prodotti')
plt.xticks(rotation=45)
plt.legend(title='Genere')
plt.tight_layout()


# Salva il grafico come file PNG
plt.savefig('colore ordinato')


plt.show()
```
